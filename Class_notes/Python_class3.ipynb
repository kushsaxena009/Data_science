```python
var = [1,2,3,43,23,34,34,23,45,54,45,45,4,5,6,6,2,3,4,1,1,9]
# output = {1:1, 2:1, 3:1, 4:1, .....]
temp = {}
print(type(var))
for i in var: #1 or 0
    if i not in temp:
        temp[i] = 1
        print("Key added",temp)
    else:
        temp[i] += 1
        print("Value added",temp)
print("Final dic",temp)
```
